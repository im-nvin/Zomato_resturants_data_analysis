```python
# Importing all the require library
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dataframe=pd.read_csv("Zomato_data.csv")
print(Dataframe)
```

```python
Dataframe
```

```python
# extracting rate values only
def handeRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
Dataframe['rate']=Dataframe['rate'].apply(handeRate)
```

```python
Dataframe
```

```python
Dataframe.info()
```

```python
Dataframe.head()
```

```python
# type of resturant
# bar graph for type of resturant
sns.countplot(x=Dataframe['listed_in(type)'])
plt.xlabel('Type of Resturant',color="Blue", fontsize=20)
plt.ylabel('count',color="Blue", fontsize=20)
```

```python
# majority of resturants falls in dinning category
```

```python
Dataframe.head()
```

```python
# how many votes has each type of resturants received from customers
grouped_data = Dataframe.groupby('listed_in(type)')['votes'].sum()

# Convert grouped data to a DataFrame
result = pd.DataFrame({'votes': grouped_data})

# Plot the data
plt.plot(result.index, result['votes'], color="green", marker="o")

# Set the labels
plt.xlabel("Type of restaurant", color="red", fontsize=20)
plt.ylabel("Votes", color="red", fontsize=20)
```

```python
# Dinning resturants has received maximum votes
```

```python
Dataframe.head()
```

```python
# what are the rating that majority of resturants has received
plt.hist(Dataframe['rate'],bins=5)
plt.title("Rating Distribution",c="Blue",size=20)
plt.show()
```

```python
# majority of resturants gets ratting between 3.75-4
```

```python
couple_data=Dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
# the majority of couples prefer to visit resturant with average price of 300
```

```python
Dataframe.head()
```

```python
# which mode has received the maximum rating online or offline
plt.figure(figsize=(6,6))
sns.boxplot(x="online_order",y="rate",data=Dataframe)
```

```python
# online order received higher rating as compare to offline rating
```

```python
Dataframe.head()
```

```python
# which type of resturant recieve more offline order so that zomato can pay to customers for switch to onine.
pivot_table=Dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title('Heatmap')
plt.xlabel('online order')
plt.ylabel('listed_in(type)')
plt.show()
```

```python
# dinning resturants recieve more offline order while cafes receive more online order
```
